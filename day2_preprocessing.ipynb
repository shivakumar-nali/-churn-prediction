```python
# DAY 2: Preprocessing & Encoding
import pandas as pd
import numpy as np

# Load cleaned data
df = pd.read_csv("data/cleaned_churn_data.csv")

# Show data types
df.dtypes
```

```python
# Some columns need conversion:
# 'TotalCharges' should be numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Check again
df.dtypes
```

```python
# Handle missing TotalCharges (became NaN)
missing_total = df["TotalCharges"].isna().sum()
print(f"✅ Missing TotalCharges: {missing_total}")

# Replace with median
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
# Drop customerID column (not predictive)
df = df.drop("customerID", axis=1)

# Confirm columns
df.columns
```

```python
# Encode categorical variables
# First, see which are object type:
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
print("✅ Categorical columns:", categorical_cols)
```

```python
# 'Churn' is the target
# Encode Churn: Yes=1, No=0
df["Churn"] = df["Churn"].map({"Yes":1, "No":0})

# Encode other categorical with pd.get_dummies
df_encoded = pd.get_dummies(df, drop_first=True)

print("✅ Encoded DataFrame shape:", df_encoded.shape)
```
